```python
_d = var('d', latex_name='\delta')
_f = var('f')

def exhibit(f):
    show(f)
    print(f)
    print("------\n")
    
def deriveAndExhibit(f,x):
    d = derivative(f,x)
    exhibit(d)
    return d

def exhibitDerivatives(f, xs, l, i=1, derivatives=[]):
    if (i > l):
        return
    
    for x in xs:
        derivs = derivatives + [x]
        down = _d
        for d in derivs:
            down = down*d
            
        show((_d^(i+1)*_f)/down)
        df = deriveAndExhibit(f, x)
        exhibitDerivatives(df, xs, l, i+1, derivs)
```

```python
x1 = var('x1')
x2 = var('x2')
f1 = -e ^ (-x1^2 - (x1^2 - x2)^2)

exhibit(f1)
exhibitDerivatives(f1, [x1,x2], 2)
```
